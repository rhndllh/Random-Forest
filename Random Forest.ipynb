```python
%matplotlib inline
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import os
import matplotlib.pylab as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import sklearn.metrics
from sklearn import datasets
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('Data/tree_addhealth.csv')
df
```

```python
print(df.shape)
df.head(10)
```

```python
#
df.isnull().sum()
```

```python
#Membuang data yang terdapat data kosong
df_clean = df.dropna()
df_clean.describe()
```

```python
print(df_clean.shape)
```

```python
#Cek Missing data kemmudian dijumlahkan
df_clean.isnull().sum()
```

```python
#Random tree hanya bisa mengolah data numeric
#Menentukan Explanatory dan Target variabel
#Merubah data string menjadi numerical agar bisa diolah random forest
num_of_numerical_cols = df_clean._get_numeric_data().columns.shape[0]
print(num_of_numerical_cols, 'numerical columns')
name_of_numerical_cols = df_clean._get_numeric_data(). columns
print(name_of_numerical_cols)
```

```python
X = df_clean[['BIO_SEX', 'HISPANIC', 'WHITE', 'BLACK', 'NAMERICAN', 'ASIAN', 'age',
             'ALCEVR1', 'ALCPROBS1', 'marever1', 'cocever1', 'inhever1', 'cigavail', 'DEP1', 'ESTEEM1', 'VIOL1',
             'PASSIST', 'DEVIANT1', 'SCHCONN1', 'GPA1', 'EXPEL1', 'FAMCONCT', 'PARACTV', 'PARPRES']]

y = df_clean.TREG1

#test_size 40% data testing 60% data training
X_train, X_test, y_train, y_test, = train_test_split(X,y, test_size=0.4, random_state = 42)
```

```python
print('X_train = ', X_train.shape)
print('X_test = ', X_test.shape)
print('y_train = ', y_train.shape)
print('y_test = ', y_test.shape)
```

```python
#Membuat Random Forest decision tree sebanyak 20 percobaan
classifier=RandomForestClassifier(n_estimators=20, random_state = 0)
classifier.fit(X_train, y_train)

predictions=classifier.predict(X_test)
```

```python
#Confusion matriks adalah tools untuk melekukan evaluasi dari hasil prediksi model
confusion_array = sklearn.metrics.confusion_matrix(y_test,predictions)
print(confusion_array)
```

```python
print('TN = ', confusion_array[0,0])
print('FN = ', confusion_array[1,0])
print('TP = ', confusion_array[1,1])
print('FP = ', confusion_array[0,1])
```

```python
classifier.score(X_test, y_test)
```

```python
sklearn.metrics.accuracy_score(y_test, predictions)
```

```python
print(classification_report(y_test, predictions))
```
